Supervised learning, unsupervised learning, and reinforcement learning differ in their approaches and applications. Supervised learning is ideal for scenarios where labeled data is available, as it focuses on mapping inputs to known outputs. It is commonly used in tasks like image classification, spam detection, and language translation. Unsupervised learning, on the other hand, is used when data lacks predefined labels. It excels in discovering hidden patterns, such as customer segmentation or anomaly detection. Reinforcement learning is fundamentally different, focusing on learning optimal actions through interaction with an environment to maximize rewards. This approach is particularly suited for applications like game AI, robotics, and self-driving cars, where decision-making in dynamic environments is crucial.

When comparing machine learning and deep learning, the key difference lies in complexity and scope. Machine learning, as a broader field, uses algorithms like decision trees or SVMs to process structured data, often requiring manual feature engineering. In contrast, deep learning, a subset of machine learning, leverages neural networks with multiple layers to automatically extract features, making it more suitable for unstructured data like images, audio, and text. Machine learning is often sufficient for smaller datasets and simpler tasks such as fraud detection or predictive maintenance, whereas deep learning excels in more complex tasks, such as real-time language translation, facial recognition, and autonomous systems, albeit with higher computational requirements.

From these distinctions, it is clear that each approach has unique strengths. Supervised learning is the go-to method for tasks requiring accuracy and labeled data, while unsupervised learning is invaluable for extracting insights from unlabeled datasets. Reinforcement learning offers a powerful framework for solving sequential decision-making problems. Machine learning’s versatility makes it widely applicable, while deep learning’s power lies in its ability to handle complex, unstructured data. Together, these techniques empower advancements in diverse fields, from healthcare and finance to robotics and artificial intelligence.

**Write a function to check whether the number is Palindrome or not.**  
- Hopefully who have did this in c++ in yor CS101

```python
def is_palindrome(number):
    # Convert the number to a string
    str_num = str(number)

    # Check if the string is equal to its reverse
    return str_num == str_num[::-1]

# Example usage
num = 121
if is_palindrome(num):
    print(f"{num} is a palindrome.")
else:
    print(f"{num} is not a palindrome.")
```


Create a Pandas DataFrame from a dictionary containing three columns: 'Name', 'Age', and 'City'. Print the DataFrame.
filter the rows where the 'Age' is greater than 30 and print the result.
Add a new column 'Salary' to the DataFrame created in Exercise 1 with values [50000, 60000, 70000]. Print the updated DataFrame
Group the DataFrame by the 'City' column and calculate the average 'Age' for each city. And plot appropriate/required graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a DataFrame from a dictionary
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 35, 45],
    "City": ["New York", "Los Angeles", "New York"]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Step 2: Filter rows where 'Age' > 30
filtered_df = df[df['Age'] > 30]
print("\nFiltered DataFrame (Age > 30):")
print(filtered_df)

# Step 3: Add a new column 'Salary'
df['Salary'] = [50000, 60000, 70000]
print("\nDataFrame with 'Salary' column:")
print(df)

# Step 4: Group by 'City' and calculate the average 'Age'
grouped_avg_age = df.groupby('City')['Age'].mean()
print("\nAverage Age by City:")
print(grouped_avg_age)

# Step 5: Plot graphs
# Bar chart for average age by city
grouped_avg_age.plot(kind='bar', color='skyblue', title='Average Age by City')
plt.xlabel('City')
plt.ylabel('Average Age')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatter plot for Age vs Salary
plt.scatter(df['Age'], df['Salary'], color='green')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.tight_layout()
plt.show()
```
